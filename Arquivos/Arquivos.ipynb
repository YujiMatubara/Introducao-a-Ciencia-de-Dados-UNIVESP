```python
import pandas as pd
import csv
```

```python
arquivo = open('MachadodeAssis.txt', 'r')
conteudo = arquivo.readlines()
arquivo.close()
```

```python
print(conteudo)
```

```python
for linha in conteudo:
    print(linha)
```

```python
#alterar em todo o texto "Rio de Janeiro" por "São Paulo"
x = ''
for linha in conteudo:  #conteudo é uma lista de strings, para usar replace precisa ser uma string
    x += linha

x = x.replace('Rio de Janeiro', 'São Paulo')    #nao altera o conteudo original
print(x)
```

```python
arquivoNovo = open('machadodeAssis2.txt', 'w')
arquivoNovo.write(x)
arquivoNovo.close()
```

```python
with open('KaggleTesla.csv', 'r') as arquivo:   #quando usa o with, nao precisa fechar o arquivo, ele fecha sozinho no final
    leitor = csv.reader(arquivo, delimiter = ',')
    for linha in leitor:
        data = linha[0]
        fechamento = linha[4]
        volume = linha[5]
        print(data, fechamento, volume)
```

```python
arquivo = 'https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv'
dados = pd.read_csv(arquivo)
dados
```

```python
plotar = dados.winner_hand.value_counts().sort_values() #u é indefinido, r é destro, l é canhoto
```

```python
plotar.plot(kind = 'bar')
```
