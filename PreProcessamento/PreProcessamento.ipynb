# Pré-Processamento de Dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv')
dados.head(5)
```

```python
dados.info()
```

```python
dados.describe(include = 'all')
```

```python
dados.isnull()
```

```python
dados.isnull().sum()
```

```python
dados['winner_entry'].describe()
```

```python
dados['winner_entry'].unique()
```

```python
dados.dropna(subset = ['loser_rank'], inplace = True)   #apagar linhas que o loser_rank é nulo, pois são poucas e influenciam no resultado
dados.dropna(axis = 1, how = 'all') #apagar colunas que todos os valores são nulos, "axis = 1" significa colunas
```

```python
dados['loser_entry'].fillna(value = 'X', inplace = True)    #substitui os valores nulos por 'X'
dados['loser_entry'].unique()
```

```python
dados['w_ace'].unique() #quantas vezes o jogador fez ace
```

```python
dados['loser_ht'].unique()  #altura do jogador
```

```python
dados.fillna(value = {'w_ace':0, 'loser_ht':dados['loser_ht'].mean()}, inplace = True) #substitui os valores nulos por 0 e pela média da altura dos jogadores
dados['w_ace'].unique()
```

```python
dados['loser_ht'].unique()
```

```python
dados['minutes'][1330:1350]
```

```python
dados['minutes'].fillna(method = 'ffill', inplace = True)   #alterar o valor NaN pelo valor da tupla anterior
dados['minutes'][1330:1350]
```

```python
dados.duplicated().sum()    #não há tuplas duplicadas
```

Para apagar duplicados de uma coluna basta:\
dados.drop_duplicates(subset = ['nome_do_campo'], inplace = True)

```python
dados2 = pd.read_csv('AAPL.csv')
dados2.head(5)
```

```python
dados2['close_ontem'] = dados2['Close'].shift(1)  #pega o dado da tupla anterior e joga para a tupla atual
dados2
```

```python
dados2['alteracao %'] = ((dados2['Close'] / dados2['close_ontem']) - 1) * 100
dados2['alteracao %']
```

```python
dados2.dropna(subset = ['alteracao %'], inplace = True) #apaga a tupla com valor NaN em alteracao %
dados2
```
