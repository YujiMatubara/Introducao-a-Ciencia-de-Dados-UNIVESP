# MatPlotLib e Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9] #gerandos pontos x e y para trabalhar
y = [1, 2, 3, 4, 2, 6, 7, 8, 9]
```

```python
plt.scatter(x, y)   #scatter é o gráfico de dispersão
plt.show()
```

```python
x1 = np.arange(-100, 100, 1) #gerando pontos no intervalo [-100, 100[ com intervalo de 1
x1
```

```python
plt.plot(x1, x1**2)  #plotando o gráfico f(x) = x²
plt.show()
```

```python
plt.plot(x1, (x1**2) - 2000)  #fazer a mesma coisa mas cortando o eixo x
plt.show()
```

```python
dias = np.arange(1, 31)
dias
```

```python
vacinados = np.random.randint(0, 1000, 30) #situação hipotética de numero de vacinados por dia
contagios = np.random.randint(0, 700, 30) #situação hipotética de numero de contagios por dia
```

```python
plt.style.use('classic')    #estilo do gráfico
plt.figure(figsize=(10, 5)) #tamanho do gráfico
plt.bar(dias, vacinados)    #plotando o gráfico de barras de vacinados
plt.plot(dias, contagios, 'r')   #plotando o gráfico de linhas de contagios em vermelho
plt.ylabel('Vacinados por dia') #legenda do eixo y
plt.show()
```

```python
dados = pd.DataFrame(dias, columns=['Dias']) #criando um dataframe com os dias
dados['Vacinados'] = vacinados
dados['Contagios'] = contagios
dados
```

```python
dados.plot(kind = 'bar', x = 'Dias', y = 'Vacinados') #plotando com dataframe
#o pandas usa o matplotlib como base para plotar os gráficos
```

```python
sns.barplot(data = dados, x = 'Dias', y = 'Vacinados') #plotando com seaborn
sns.lineplot(data = dados, x = 'Dias', y = 'Contagios', color = 'red')
```
