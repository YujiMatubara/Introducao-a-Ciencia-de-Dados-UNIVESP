```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
minutospordia = np.random.randint(0, 60, 30)
minutospordia
```

```python
ax = sns.histplot(minutospordia)
ax.set(xlabel = 'Minutos', ylabel = 'Frequência', title = 'Exercícios Diários')
plt.show()
```
