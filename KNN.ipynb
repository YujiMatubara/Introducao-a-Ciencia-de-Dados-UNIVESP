# KNN
K-Nearest Neighbors

```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
iris.target_names
```

```python
df['target_name'] = iris.target_names[df.target]
df
```

```python
iris_features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
iris_features
```

```python
y = df.target
y
```

```python
X = df[iris_features]
X
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
modelo = KNeighborsClassifier(3)    #treino
modelo.fit(X, y)
```

```python
print(modelo.predict(X))    #teste
```

```python
modelo.score(X, y)  #acurácia do teste
```
